```python
import numpy as np
import pandas as pd
import scanpy as sc
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=96)
```

```python
adata = sc.read_h5ad('./lung_dev_covid_052620.h5ad')
```

```python
adata
```

```python
sc.pl.umap(adata, color=['leiden'])
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 6,3
ax = sc.pl.violin(adata, ['Bsg'], groupby='leiden', rotation=90)
```

```python
at1 = adata[adata.obs['leiden'].isin(['AT1'])]
at2 = adata[adata.obs['leiden'].isin(['AT2'])]
secretory = adata[adata.obs['leiden'].isin(['Secretory'])]
ciliated = adata[adata.obs['leiden'].isin(['Ciliated'])]
```

```python
ax = sc.pl.violin(at1, ['Bsg'], groupby='timepoint', swapaxes=True)
ax = sc.pl.violin(at2, ['Bsg'], groupby='timepoint', swapaxes=True)
ax = sc.pl.violin(secretory, ['Bsg'], groupby='timepoint', swapaxes=True)
ax = sc.pl.violin(ciliated, ['Bsg'], groupby='timepoint', swapaxes=True)
```
